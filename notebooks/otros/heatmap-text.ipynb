## Plotting heatmaps with text

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# Make an array with ones in the shape of an 'X'
a = np.eye(10,10)
a += a[::-1,:]

fig = plt.figure()
ax1 = fig.add_subplot(121)
# Bilinear interpolation - this will look blurry
ax1.imshow(a, interpolation='bilinear', cmap=plt.cm.Greys_r)

ax2 = fig.add_subplot(122)
# 'nearest' interpolation - faithful but blocky
ax2.imshow(a, interpolation='nearest', cmap=plt.cm.Greys_r)

plt.show()
```

```python
import seaborn as sns

tokens = ["this", "is", "something", "to", "test"]

X = 0.01 * np.random.randn(30, len(tokens))

sns.heatmap(
    X,
    xticklabels=tokens,
)
```

## With PlotLy

```python
import plotly.graph_objects as go
import nltk
from nltk.tokenize import wordpunct_tokenize
text = "Lorem ipsum dolor sit amet, consectetur adipiscing elit. Praesent tempor ipsum vel neque gravida porttitor. Proin libero diam, tristique a tempus sit amet, vulputate id nibh."

tokens = wordpunct_tokenize(text)
print(tokens)
X = np.random.randn(150, len(tokens))

fig = go.Figure(
    data=go.Heatmap(
        z=X,
        x=tokens,
))
fig.update_layout(
    autosize=False,
    width=1200,
    height=2500,
)

fig.show()
```
