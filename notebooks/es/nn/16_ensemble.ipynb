```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import pandas as pd
import csv

df_dev = pd.read_table("../../../data/es/dev_es.tsv", index_col="id", quoting=csv.QUOTE_NONE)
df_train = pd.read_table("../../../data/es/train_es.tsv", index_col="id", quoting=csv.QUOTE_NONE)
df_test = pd.read_table("../../../data/es/test_es.tsv", index_col="id", quoting=csv.QUOTE_NONE)

print("Instancias de entrenamiento: {}".format(len(df_train)))
print("Instancias de desarrollo: {}".format(len(df_dev)))
print("Instancias de test: {}".format(len(df_test)))
```

Hagamos tokenize sobre el texto

## Bow Model

```python
from hate.nn.preprocessing import Tokenizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

tok_args = {
    "preserve_case": False,
    "deaccent": False,
    "reduce_len": True,
    "strip_handles": False,
    "alpha_only": False,
    "stem": True
}

tokenizer = Tokenizer(**tok_args)


vect = CountVectorizer(tokenizer=tokenizer.tokenize, 
                       max_df=0.65, min_df=0.00075, ngram_range=(1, 2), binary=True)

vect.fit(df_train["text"])

bow_train = vect.transform(df_train["text"])
bow_dev = vect.transform(df_dev["text"])

bow_train.shape, bow_dev.shape
```

## Char Input

```python
from keras.preprocessing.text import Tokenizer as KerasTokenizer
vocab_size = 200

tokenize_args = {
    "preserve_case": False, 
    "deaccent": True,
    "reduce_len": True, 
    "strip_handles": False,
    "stem": True,
    "alpha_only": False
}

tokenizer = Tokenizer(**tokenize_args)

text_train = [" ".join(tokenizer.tokenize(t)) for t in df_train["text"].values]
text_dev = [" ".join(tokenizer.tokenize(t)) for t in df_dev["text"].values]

char_tokenizer = KerasTokenizer(num_words=vocab_size, char_level=True, 
                           lower=tokenize_args["preserve_case"])
char_tokenizer.fit_on_texts(text_train)

X_char_train = char_tokenizer.texts_to_sequences(text_train)
X_char_dev = char_tokenizer.texts_to_sequences(text_dev)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


plt.rcParams["figure.figsize"] = (10,3)


plt.subplot(1, 2, 1)
plt.hist([len(seq) for seq in X_char_train])


plt.subplot(1, 2, 2)
plt.hist([len(seq) for seq in X_char_dev])
```

## Modelo múltiple inputs
Hagamos un primer modelo con múltiple inputs.

```python
%%capture
from elmoformanylangs import Embedder

e = Embedder("../../../models/elmo/es/")
```

```python
from keras.models import Model
from keras.layers import Concatenate, Dense, CuDNNLSTM
from keras.optimizers import Adam
from keras.utils import plot_model
from hate.nn import CharModel, ElmoModel, BowModel, MergeModel

dropout = [0.80, 0.55]

char_model = CharModel(max_charlen=max_charlen, vocab_size=200, recursive_class=CuDNNLSTM, dropout=dropout)
elmo_model = ElmoModel(50, embedder=e, recursive_class=CuDNNLSTM, dropout=dropout)
bow_model = BowModel(
    num_words=5500, vectorize_args={"max_df":0.65, "min_df": 0.00075}, dropout=dropout
)

merge_model = MergeModel([char_model, elmo_model, bow_model])

optimizer_args = {
    "lr": 0.0003,
    "decay": 0.01,
    "amsgrad": False
}


merge_model.compile(loss='binary_crossentropy', 
              optimizer=Adam(**optimizer_args), 
              metrics=['accuracy'])


from IPython.display import Image

print(merge_model.summary())


plot_model(merge_model, to_file='model.png')
Image('model.png')
```

```python
X_train, y_train = df_train["text"], df_train["HS"]
X_dev, y_dev = df_dev["text"], df_dev["HS"]
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

checkpointer = ModelCheckpoint('/tmp/ensemble_16.h5', save_best_only=True, monitor='val_acc', verbose=1)
early_stopper = EarlyStopping(monitor='val_loss', patience=15)
history = merge_model.fit(X_train, y_train, callbacks=[checkpointer, early_stopper],
          validation_data=(X_dev, y_dev), epochs=300, batch_size=48)
```

```python
from hate.utils import print_evaluation

merge_model.load_weights(checkpointer.filepath)


print_evaluation(merge_model, X_dev, y_dev)
```
